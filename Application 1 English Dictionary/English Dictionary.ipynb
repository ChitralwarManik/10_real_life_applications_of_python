
*If the user has spelling mistake of an word then our program should match the spelling to the correct word and also suggest the currect word to the user. for this we use a liabrary called **get_close_matches** from difflib.*  

```python
from difflib import get_close_matches
```

*Loading json data*

```python
data = json.load(open("Dictionary_data.json"))
```

*Define a function called **translate** that will do following things:*
1. if the user has entered the word in lower or upper or both the cases the program will convert it into smaller case.
2. Then it will search the word in a pattern, if the user enteres the word is found in data then it will return that, else it not found it then it will convert the word into title format then search it, if that also not found then it will convert the word into upper case then it will check.
3. If the user has entered wrong spelling of an word then it will use the get_close_matches liabrary to match the similar word and suggest to the user to varify.
4. If the user entered word is not found in the data, then it will print "The word doesn't exist. Please double check it."


```python
def translate(w):
    w = w.lower()
    if w in data:
        return data[w]
    elif w.title() in data:
        return data[w.title()]
    elif w.upper() in data: #in case user enters words like USA or NATO
        return data[w.upper()]
    elif len(get_close_matches(w, data.keys())) > 0:
        yn = input("Did you mean %s instead? Enter Y if yes, or N if no: " % get_close_matches(w, data.keys())[0])
        if yn == "Y":
            return data[get_close_matches(w, data.keys())[0]]
        elif yn == "N":
            return "The word doesn't exist. Please double check it."
        else:
            return "We didn't understand your entry."
    else:
        return "The word doesn't exist. Please double check it."
word = input("Enter word: ")
output = translate(word)
if type(output) == list:
    for item in output:
        print(item)
else:
    print(output)
```
